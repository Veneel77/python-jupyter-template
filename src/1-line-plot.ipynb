### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

x_values = list(range(11))  # X values from 0 to 10
y_values = [x**2 for x in x_values]  # Y values as the square of X

plt.plot(x_values, y_values, marker='o', linestyle='-')  # Plot with markers
plt.title("Square Function Line Plot")  # Title
plt.xlabel("X Values")  # X-axis label
plt.ylabel("Y = X^2")  # Y-axis label
plt.grid(True)  # Add grid
plt.show()  # Display the plot
```
