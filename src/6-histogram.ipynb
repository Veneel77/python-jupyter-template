### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

# Function to create a histogram with enhanced error handling
def plot_histogram(data, bins=5, color='blue'):
    try:
        # Validate input data
        if not isinstance(data, list) or not all(isinstance(i, (int, float)) for i in data):
            raise TypeError("'data' must be a list of numerical values.")
        
        # Validate bins: Allow both positive integers and valid string options for bins
        valid_bin_options = ['auto', 'fd', 'doane', 'scott', 'rice', 'sturges', 'sqrt']
        if not (isinstance(bins, int) and bins > 0) and not (isinstance(bins, str) and bins in valid_bin_options):
            raise ValueError(f"'bins' must be a positive integer or one of {valid_bin_options}.")

        # Create a histogram
        plt.hist(data, bins=bins, color=color, edgecolor='black')

        # Labeling the axes and the plot
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.title('Histogram')

        # Show the plot
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Example usage
if __name__ == "__main__":
    data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
    plot_histogram(data, bins='auto', color='green')  # Now supports 'auto' and other bin strategies
```
