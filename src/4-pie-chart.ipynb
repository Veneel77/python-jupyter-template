### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

def plot_pie_chart(labels, sizes):
    try:
        # Validate input types
        if not isinstance(labels, list) or not isinstance(sizes, list):
            raise TypeError("Both 'labels' and 'sizes' should be lists.")
        
        # Validate that both lists have the same length
        if len(labels) != len(sizes):
            raise ValueError("'labels' and 'sizes' must have the same length.")
        
        # Ensure all sizes are numerical
        if not all(isinstance(size, (int, float)) for size in sizes):
            raise TypeError("All elements in 'sizes' must be numerical.")
        
        # Create a pie chart
        plt.figure(figsize=(7, 7))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'orange'])
        
        # Title
        plt.title('Programming Language Usage')
        
        # Show the plot
        plt.show()
    
    except Exception as e:
        print(f"Error: {e}")

# Example usage
if __name__ == "__main__":
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    plot_pie_chart(labels, sizes)
```
