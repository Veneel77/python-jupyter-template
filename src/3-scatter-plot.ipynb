### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Function to create a scatter plot with annotations and error handling
def plot_scatter(x, y):
    try:
        # Validate input types
        if not isinstance(x, list) or not isinstance(y, list):
            raise TypeError("Both 'x' and 'y' should be lists.")
        
        # Validate that both lists have the same length
        if len(x) != len(y):
            raise ValueError("'x' and 'y' must have the same length.")
        
        # Ensure all values are numerical
        if not all(isinstance(i, (int, float)) for i in x + y):
            raise TypeError("All elements in 'x' and 'y' must be numerical.")
        
        # Create a scatter plot
        plt.figure(figsize=(8, 5))  # Set figure size
        plt.scatter(x, y, color='blue')
        
        # Annotate each point with its (x, y) value
        for i in range(len(x)):
            plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')
        
        # Labeling the axes and the plot
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Scatter Plot with Annotations')
        
        # Show the plot
        plt.show()
    
    except Exception as e:
        print(f"Error: {e}")

# Example usage
if __name__ == "__main__":
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]
    plot_scatter(x, y)
```
