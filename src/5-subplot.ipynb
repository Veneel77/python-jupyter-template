### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

def validate_data(x, y, categories, values, labels, sizes):
    try:
        if not (isinstance(x, list) and isinstance(y, list) and isinstance(categories, list) and isinstance(values, list) and isinstance(labels, list) and isinstance(sizes, list)):
            raise TypeError("All inputs must be lists.")
        
        if len(x) != len(y):
            raise ValueError("'x' and 'y' lists must have the same length.")
        
        if len(categories) != len(values):
            raise ValueError("'categories' and 'values' must have the same length.")
        
        if len(labels) != len(sizes):
            raise ValueError("'labels' and 'sizes' must have the same length.")
        
        if not all(isinstance(i, (int, float)) for i in y + values + sizes):
            raise TypeError("All numerical lists must contain only numbers.")
        
        if not all(isinstance(i, str) for i in categories + labels):
            raise TypeError("All category and label names must be strings.")
        
        return True
    except Exception as e:
        print(f"Error: {e}")
        return False

# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]

if validate_data(x, y, categories, values, labels, sizes):
    # Create a 2x2 grid of subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))

    # First subplot: Line plot
    axs[0, 0].plot(x, y, marker='o', linestyle='-')
    axs[0, 0].set_title('Line Plot')
    axs[0, 0].grid(True)

    # Second subplot: Bar plot
    axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'orange'])
    axs[0, 1].set_title('Bar Plot')

    # Third subplot: Scatter plot
    axs[1, 0].scatter(x, y, color='purple', edgecolors='black')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].grid(True)

    # Fourth subplot: Pie chart
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['cyan', 'magenta', 'yellow', 'gray'])
    axs[1, 1].set_title('Pie Chart')

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```
