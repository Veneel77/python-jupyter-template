### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt
import random

# Function to plot a bar chart with error handling
def plot_bar_chart(categories, values):
    try:
        # Validate input types
        if not isinstance(categories, list) or not isinstance(values, list):
            raise TypeError("Both 'categories' and 'values' should be lists.")
        
        # Validate that both lists have the same length
        if len(categories) != len(values):
            raise ValueError("'categories' and 'values' must have the same length.")
        
        # Ensure all values are numerical
        if not all(isinstance(v, (int, float)) for v in values):
            raise TypeError("All elements in 'values' must be numerical.")
        
        # Ensure all categories are strings
        if not all(isinstance(c, str) for c in categories):
            raise TypeError("All elements in 'categories' must be strings.")
        
        # Dynamically generate colors for bars
        colors = ["#" + "".join(random.choices("0123456789ABCDEF", k=6)) for _ in categories]
        
        # Create bar plot
        plt.figure(figsize=(8, 5))  # Set figure size for better visibility
        plt.bar(categories, values, color=colors)
        
        # Labeling the axes and the plot
        plt.xlabel('Categories')
        plt.ylabel('Values')
        plt.title('Bar Plot with Different Colors')
        
        # Show the plot
        plt.show()
    
    except Exception as e:
        print(f"Error: {e}")

# Example usage
if __name__ == "__main__":
    categories = ['A', 'B', 'C', 'D', 'E', 'F']  # Extended categories
    values = [5, 7, 3, 9, 4, 6]  # Matching values
    plot_bar_chart(categories, values)


    
```
